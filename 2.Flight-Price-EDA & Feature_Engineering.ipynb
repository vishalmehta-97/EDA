## Feature Engineering

```python
data1["Date"]=data1["date"].str.split('-').str[0]
data1["Month"]=data1["date"].str.split('-').str[1] 
data1["Year"]=data1["date"].str.split('-').str[2] 
## with the help of this we have splitted the date in a single column
```

```python
data1.head(2)
data1.info()  ## Here we can see that after seprating the dates they are still in object type so now we'll convert this in to integer
```

```python
data1["Date"]=data1["Date"].astype(int)
data1["Month"]=data1["Month"].astype(int)
data1["Year"]=data1["Year"].astype(int)
```

```python
data1.info()
```

```python
data1=data1.drop('date',axis=1)
```

```python
data1.head(2)
```

```python
data1['Arrival_hour']=data1["arr_time"].str.split(':').str[0]
data1['Arrival_min']=data1["arr_time"].str.split(':').str[1]
data1['Arrival_hour']=data1['Arrival_hour'].astype(int)
data1['Arrival_min']=data1['Arrival_min'].astype(int)
data1.info()
```

```python
data1=data1.drop("arr_time",axis=1)
```

```python
data1.head(2)
```

```python
data1['Departure_hour']=data1["dep_time"].str.split(':').str[0]
data1['Departure_min']=data1["dep_time"].str.split(':').str[1]
data1['Departure_hour']=data1['Departure_hour'].astype(int)
data1['Departure_min']=data1['Departure_min'].astype(int)
data1=data1.drop("dep_time",axis=1)
data1.info()
```

```python
data1.head(2)
```

```python
data1["stop"].unique()   ## in this dataset the values were very odd so i did this to make this into a valuable feature
data1["stop"]=data1["stop"].str.split('-').str[0]
data1.head(3)
```

```python
data1["stop"].unique()
```

```python
data1["stop"]=data1["stop"].map({'non':0,'1':1,'2+':2})
data1.head(2)
```

```python
data2.head(2)   ## now to work on time_taken Feature i am going to use this dataset4
```

```python
data2["Time_taken(hrs)"]=data2["time_taken"].str.split(' ').str[0].str.split('h').str[0]
# data2["Time_taken(hrs)"]=data2["Time_taken(hrs)"].astype(int)
data2["Time_taken(hrs)"].unique()
```

```python
mask=data2["Time_taken(hrs)"].str.contains(r"\.", na=False)    ## there were some decimal values in the dataset that's why did this type of split in there so we can get the whole Hour value for our new column
data2.loc[mask,"Time_taken(hrs)"]=(data2.loc[mask,"Time_taken(hrs)"].str.split('.').str[0])
```

```python
data2["Time_taken(hrs)"].unique()
data2["Time_taken(hrs)"]=data2["Time_taken(hrs)"].astype(int)
```

```python
data2.head()
```

```python
data2["Time_taken(min)"]=data2["time_taken"].str.split(' ').str[1].str.split('m').str[0]
```

```python
data2["Time_taken(min)"].unique()
```

```python
mask = data2["Time_taken(min)"] == ''  ## these are the values which happen because of these 4 values which were with the wrong entry in the data in time_taken Column (like this 1.03h m)
```

```python
data2.loc[mask, "Time_taken(min)"] = (data2.loc[mask, "time_taken"]
    .str.extract(r'\.(\d+)')[0])
```

```python
data2["Time_taken(min)"]=data2["Time_taken(min)"].astype(int)
data2["Time_taken(min)"].unique()
```
